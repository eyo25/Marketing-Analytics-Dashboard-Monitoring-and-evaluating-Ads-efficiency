```python
import pandas as pd
from datetime import datetime
from telegram import Bot
import asyncio
import nest_asyncio

# Apply nest_asyncio
nest_asyncio.apply()

# Your Telegram bot token
BOT_TOKEN = 'your_bot_token'
# List of channel usernames (without @)
CHANNEL_USERNAMES = [
    'BoAEth',
    'combankethofficial',
    'Hibret Bank'
    # Add more channels as needed
]

# Initialize the bot
bot = Bot(token=BOT_TOKEN)

# Function to get subscriber count
async def get_subscriber_count(channel_username):
    try:
        chat = await bot.get_chat(f'@{channel_username}')
        return chat.get('members_count', 'Unavailable')
    except Exception as e:
        print(f"Error fetching data for channel {channel_username}: {e}")
        return 'Unavailable'

# Function to collect and save data
async def collect_data():
    data_list = []
    for username in CHANNEL_USERNAMES:
        subscriber_count = await get_subscriber_count(username)
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        data = {
            'timestamp': timestamp,
            'channel_username': username,
            'subscriber_count': subscriber_count
        }
        data_list.append(data)
    return data_list

# File path to save the CSV
csv_file_path = 'telegram_channels_data.csv'

# Function to save data to CSV
def save_to_csv(data_list, file_path):
    df = pd.DataFrame(data_list)
    try:
        df.to_csv(file_path, mode='a', header=not pd.io.common.file_exists(file_path), index=False)
    except Exception as e:
        print(f"Error saving data to CSV: {e}")

# Main function to run the async functions
async def main():
    data_list = await collect_data()
    save_to_csv(data_list, csv_file_path)
    print(f"Data saved to {csv_file_path}")

# Run the main function
if __name__ == "__main__":
    asyncio.get_event_loop().run_until_complete(main())

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

# List of Telegram channel URLs
CHANNEL_URLS = [
    'https://t.me/BoAEth',
    'https://t.me/combankethofficial',
    'https://t.me/HibretBank',
    'https://t.me/awash_bank_official',
    'https://t.me/GlobalBankEth'
    
    # Add more channels as needed
]

# Function to get subscriber count from a channel
def get_subscriber_count(channel_url):
    try:
        response = requests.get(channel_url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            subscriber_count_element = soup.find('div', {'class': 'tgme_page_extra'})
            if subscriber_count_element:
                subscriber_count_text = subscriber_count_element.get_text()
                subscriber_count = int(subscriber_count_text.split(' ')[0].replace(',', ''))
                return subscriber_count
        else:
            print(f"Error fetching data for {channel_url}: HTTP {response.status_code}")
    except Exception as e:
        print(f"Error fetching data for {channel_url}: {e}")
    return 'Unavailable'

# Function to collect and save data
def collect_data():
    data_list = []
    for url in CHANNEL_URLS:
        subscriber_count = get_subscriber_count(url)
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        data = {
            'timestamp': timestamp,
            'channel_url': url,
            'subscriber_count': subscriber_count
        }
        data_list.append(data)
    return data_list

# File path to save the CSV
csv_file_path = 'telegram_channels_data1.csv'

# Function to save data to CSV
def save_to_csv(data_list, file_path):
    df = pd.DataFrame(data_list)
    try:
        df.to_csv(file_path, mode='a', header=not pd.io.common.file_exists(file_path), index=False)
    except Exception as e:
        print(f"Error saving data to CSV: {e}")

# Main function to run the script
if __name__ == "__main__":
    data_list = collect_data()
    save_to_csv(data_list, csv_file_path)
    print(f"Data saved to {csv_file_path}")
```

```python
import pandas as pd
from google_play_scraper import app
from datetime import datetime, timedelta
import time

# Function to fetch app data
def fetch_app_data(app_id):
    try:
        result = app(app_id)
        data = {
            'app_id': app_id,
            'title': result['title'],
            'installs': result['installs'],
            'ratings_count': result['ratings'],
            'average_rating': result['score'],
            'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        return data
    except Exception as e:
        print(f"Failed to fetch data for app ID {app_id}: {e}")
        return None

# List of bank app IDs
app_ids = [
    'com.boa.boaMobileBanking',           # Bank of Abyssinia
    'com.combanketh.mobilebanking',       # Commercial Bank of Ethiopia
    'com.ekenya.awashwallet',             # Awash Bank
    'com.act.wegagen',                    # Wegagen Bank
    'com.ofss.dgbb',                      # Bank of Ethiopia (example, replace with actual ID)
]

# File path to save the CSV
csv_file_path = 'bank_apps_hourly_data.csv'

# Function to simulate fetching data for the last month at hourly intervals
def simulate_hourly_data_collection():
    end_date = datetime.now()
    start_date = end_date - timedelta(days=30)

    current_date = start_date
    while current_date <= end_date:
        for hour in range(24):
            timestamp = current_date + timedelta(hours=hour)
            data_list = []
            for app_id in app_ids:
                app_data = fetch_app_data(app_id)
                if app_data:
                    app_data['timestamp'] = timestamp.strftime("%Y-%m-%d %H:%M:%S")
                    data_list.append(app_data)
                    print(f"Fetched data for {app_data['title']} at {app_data['timestamp']}")
            df = pd.DataFrame(data_list)
            try:
                df.to_csv(csv_file_path, mode='a', header=not pd.io.common.file_exists(csv_file_path), index=False)
            except Exception as e:
                print(f"Error saving data to CSV: {e}")
        current_date += timedelta(days=1)

# Main function
if __name__ == "__main__":
    simulate_hourly_data_collection()
    print(f"Data saved to {csv_file_path}")
```
